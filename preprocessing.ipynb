# Preprocessing

# Robest scaler
- If your dataset contains more no of outliers,then robust scaling is best technique for scaling

```python
### robust scaler formula


#  xi-Q2(x)
#  ------------
# Q3(x)-Q1(x)

### Inter Quartile Range(IQR)
#  Q1->25%(Lower Quartile)
#  Q2->50%(Median Quartile)
# Q3->75%(upper quartile)
```

```python
import pandas as pd
import numpy as np
```

```python
data1=pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Advertisement.csv')
data1
```

# Standard scaler
- most commonly used scalling technique for preprocessing. amd it scales 0 to mean 

```python
### Standard scaler formula
# xi-mean(x)
# ----------
# std(x)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# step1: import model
#step2: intialize model into new object
#step2: fit the model to dataset
#step4: coinvert array values intpo dataframe
```

```python
ss=StandardScaler()
```

```python
ss2=ss.fit_transform(data1)
ss2
```

```python
ss2=pd.DataFrame(ss2,columns=data1.columns)
ss2.head()
```

```python
fig,(ax1,ax2)=plt.subplots(ncols=2,figsize=(5,5))
sns.kdeplot(data1['TV'],ax=ax1)
sns.kdeplot(data1['radio'],ax=ax1)
sns.kdeplot(data1['newspaper'],ax=ax1)
sns.kdeplot(data1['sales'],ax=ax1)
            
sns.kdeplot(ss2['TV'],ax=ax2)
sns.kdeplot(ss2['radio'],ax=ax2)
sns.kdeplot(ss2['newspaper'],ax=ax2)
sns.kdeplot(ss2['sales'],ax=ax2)
```
